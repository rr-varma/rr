```python
def menu():
  print("**********************************************************************************")
  print('''Hii ..
  What do you want me to do ?
          \033[1m1. Text Sentiment
          2. Text Summarization
          3. image to sentiment
          4. Movie Sentiment
          5. exit.\033[0m''')

  print("**********************************************************************************")


while True:
  menu()
  choice = input("Enter choice : ")
  if choice in ('1','2','3','4'):

    if choice == '1':
      print("\033[1mtext sentiment you opted..\033[0m")
      print("--------------------------------------------------------------------------------")

      from textblob import TextBlob

      # Sample text for sentiment analysis
      text = input("Enter the text or Paste here : ")
      print("-------------------------------------------------------------------------------")
      print("**sentiment**")
      # Create a TextBlob object
      blob = TextBlob(text)

      # Perform sentiment analysis
      sentiment_score = blob.sentiment.polarity

      # Interpret sentiment score
      if sentiment_score > 0:
        print("\033[3m Positive sentiment\033[0m")
      elif sentiment_score < 0:
        print("\033[3m Negative sentiment\033[0m")
      else:
        print("\033[3m Neutral sentiment\033[0m")

    elif choice == '2':
      print("\033[1m Text Summarization \033[0m")
      print("--------------------------------------------------------------------------------")
      from transformers import BartForConditionalGeneration, BartTokenizer

      def text_summarization_bart(text):

        # Load pre-trained BART model and tokenizer
        model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
        tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")

        # Tokenize input text
        inputs = tokenizer([text], max_length=1024, return_tensors="pt", truncation=True)

        # Generate summary
        summary_ids = model.generate(inputs["input_ids"], num_beams=4, min_length=20, max_length=50, early_stopping=True)
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary

      # Example usage:
      input_text = input("Enter text : ")

      print("-----------------------------------------------------------------------")
      print("**\033[3m Here is your summary \033[0m**")
      print(text_summarization_bart(input_text))

    elif choice == '3':
      import numpy as np
      from keras.preprocessing import image
      from keras.applications.vgg16 import VGG16, preprocess_input
      from keras.models import Model
      from keras.layers import Dense, GlobalAveragePooling2D
      from keras.optimizers import Adam
      from IPython.display import Image

      # Load pre-trained VGG16 model without the top (fully connected) layers
      base_model = VGG16(weights='imagenet', include_top=False)

      # Add custom top layers for sentiment classification
      x = base_model.output
      x = GlobalAveragePooling2D()(x)
      x = Dense(1024, activation='relu')(x)
      predictions = Dense(1, activation='sigmoid')(x)  # Output layer for binary sentiment classification

      model = Model(inputs=base_model.input, outputs=predictions)

      # Freeze pre-trained layers
      for layer in base_model.layers:
          layer.trainable = False

      # Compile the model
      model.compile(optimizer=Adam(lr=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

      # Define image preprocessing function
      def preprocess_image(img_path):
          img = image.load_img(img_path, target_size=(224, 224))
          display(img)
          x = image.img_to_array(img)
          x = np.expand_dims(x, axis=0)
          x = preprocess_input(x)
          return x

      # Example usage to predict sentiment for an image
      def predict_sentiment(img_path):
          preprocessed_img = preprocess_image(img_path)
          sentiment = model.predict(preprocessed_img)
          if sentiment > 0.5:
              return "Positive"
          else:
              return "Negative"

      # Example usage
      img_path = filename=input("Enter image filename: ")


      predicted_sentiment = predict_sentiment(img_path)
      print("Predicted sentiment for the image is:", predicted_sentiment)


    elif choice == '4':
      print("\033[1m Movie Sentiment \033[0m")
      import requests

      def get_movie_rating(api_key, movie_title):
          base_url = "https://api.themoviedb.org/3/search/movie"

          params = {
              'api_key': api_key,
              'query': movie_title
          }

          try:
              response = requests.get(base_url, params=params)
              response.raise_for_status()  # Raise an exception for bad responses

              data = response.json()

              if data['results']:
                  # Assuming the first result is the desired movie
                  movie_id = data['results'][0]['id']

                  # Fetching additional details, including rating
                  movie_details_url = f"https://api.themoviedb.org/3/movie/{movie_id}"
                  params = {
                      'api_key': api_key,
                      'append_to_response': 'credits'  # Include additional information like credits
                  }
                  details_response = requests.get(movie_details_url, params=params)
                  details_data = details_response.json()

                  # Extracting rating from the response
                  rating = details_data['vote_average']
                  return rating
              else:
                  return "Movie not found"
          except requests.exceptions.RequestException as e:
              return f"Error: {e}"

      # Replace 'YOUR_API_KEY' with your actual TMDb API key
      api_key = 'eb27122e2981005498200bc71800750d'
      movie_title = input("Enter the movie title: ")

      rating = get_movie_rating(api_key, movie_title)
      #print(f"The rating of {movie_title} is: {rating}")

      if rating >=7.5:
        print("Positive")
      elif rating <7.5 and rating >4.5:
        print("neutral")
      elif rating == 0:
        print(" Movie NOT Found ")
      else:
        print("negative")




  elif choice == '5':
    print("thanks for using me....")
    break

  else:
    print("invalid input ")
```
