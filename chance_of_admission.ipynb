```python
import pandas as pd
```

```python
admi = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
admi.info()
```

```python
admi.columns
```

```python
y = admi['Chance of Admit ']   #predicted value
x = admi[['Serial No', 'GRE Score', 'TOEFL Score', 'University Rating', ' SOP','LOR ', 'CGPA', 'Research']]  #features
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
#spliting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2529)
```

```python
#model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_pred,y_test)
```
